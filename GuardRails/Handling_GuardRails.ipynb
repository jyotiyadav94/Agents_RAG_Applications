```python
# This should be in a .co file

colang_content = """
# Define limits
define user ask weather
    "what's the weather like today?"
    "tell me the weather forecast."
    "is it raining outside?"

define bot answer weather
    "I'm a shopping assistant and cannot assist you with weather-related inquiries."

define flow weather
    user ask weather
    bot answer weather
    bot offer help

# Here we use the chatbot for anything else
define flow
    user ...
    $answer = execute bot_response(inputs=$last_user_message)
    bot $answer
"""
```

```python
async def func(inputs: str):
    return "Hello, I am a Shopping Assistant! How can I help you?"
```

```python
yaml_content = """
models:
- type: main
  engine: openai
  model: gpt-3.5-turbo
"""
```

```python
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = ""
```

```python
from nemoguardrails import LLMRails, RailsConfig

# Initialize rails config
config = RailsConfig.from_content(
    colang_content=colang_content,
    yaml_content=yaml_content
)
# Create rails
rails = LLMRails(config)
```

```python
rails.register_action(action=func, name="bot_response")
```

```python
response=await rails.generate_async(prompt="What is the weather like today?")
print(response)
```
